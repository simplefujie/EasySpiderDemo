# 一、使用Python自带的urllib库爬虫
## 1.1 先爬取百度网站，看看能不能爬取到信息，如果可以爬到东西，看看我们具体爬到了什么。

```python
import urllib.request

url="https://www.baidu.com/"
response=urllib.request.urlopen(url)
print(response)
```

## 1.2 读取并分析爬取到的内容
从返回的打印出来的信息表明，我们通过urllib.request.urlopen(url)这个方法，确实爬取到了信息，这个信息被保存在0x00000225CA2295C8这个地方。接下来我们将其打印出来，看看它具体是什么。

```python
data=response.read().decode()
print(data)
```

从打印出来的信息我们可以发现，我们得到了百度网站的首页的html信息，说明我们确实通过urllib.request.open(url)这个指令，爬取到了百度网的信息。<br/>
接下来介绍一下我们所用函数的作用:
- response=urllib.request.urlopen(url)：通过这个方法，获取到网页的信息，保存到内存中去
- data=response.read().decode()：从内存0x00000225CA2295C8中读取信息，并且解码为我们可以认识的样子

# 二、使用第三方库requests爬虫
## 2.1 使用requests爬取百度网的信息
首先引入requests模块，看看我们是否能够访问百度

```python
import requests
response=requests.get(url)
print(response)
```

显示200，说明我们成功访问到了百度，与百度建立了连接
## 2.2 读取分析爬取到的内容
从response中，读取出来具体的网页的内容

```python
data=response.text
print(data)
```

可以看到，我们直接使用response.text，就直接得到了完整的html网页，可见使用第三方库requests来爬虫，大大减少了我们的代码量。<br/>
下面总结一下使用requests爬虫的两步
- 使用response=requests.get(url)获取网页信息
- 使用data=response.text读取网页信息

# 三、什么是headers
现在很多网站因为诸多原因，屏蔽了来自网络爬虫的访问，所以这个时候我们爬虫时必须模仿浏览器的行为。在我们爬虫的时候，在爬虫模块中加入headers信息，这样服务器就以为我们是浏览器发送的访问，而不是爬虫。这样我们就可以顺利爬虫给了。

## 3.1 不能直接爬取哔哩哔哩
下面展示爬取哔哩哔哩时的情况，由于哔哩哔哩设置了反爬虫机制，如果我们不加入headers，那么我们将不能直接访问哔哩哔哩

```python
url="https://www.bilibili.com/"
response=urllib.request.urlopen(url)
data=response.read().decode()
print(data)
```

可以看到，我们直接爬取哔哩哔哩时，报出了HTTPError这个错误，说明直接爬取哔哩哔哩是不可以的，下面我们引入headers
## 3.2 在urllib中添加headers以爬取哔哩哔哩
### 首先介绍一下headers是什么
headers是浏览器在访问服务器时，发送给服务器的一行字符串，指明浏览器的信息。这样服务器就可以识别到浏览器的型号。
### 怎样获取headers
获取headers的方法有很多，简单一点的就是直接百度，会有别人分享的headers信息，直接拿来用就行了。也可以手动根据自己的浏览器信息来获取headers，下面以chrome浏览为例子
- 打开任意网页，按F12进入浏览器中的调试台
- 点击调试台中的Network后，按F5刷新页面
- 这个时候浏览器会获取到很多文件，点击任意一个
- 复制Headers栏下的user.agent，这个就是我们需要的headers信息
如下所示：
![GYbmp4.png](https://s1.ax1x.com/2020/04/02/GYbmp4.png)

```python
# 展示在urllib中使用headers的方法
headers={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36"
}
req=urllib.request.Request(url,headers=headers) # 向服务器提供headers信息，以建立连接
response=urllib.request.urlopen(req) # 从服务器读取数据
data=response.read().decode() # 读取数据并且展示
print(data)
```

可以看到，在urllib中添加到headres之后，我们就可以访问到哔哩哔哩了
### 3.3 使用requests库爬虫时，添加headers
首先，直接使用requests爬取哔哩哔哩是能够直接访问的，可能是因为第三方库requests本身继承了对浏览器的模仿。如下所示，我们没有添加headers还是可以使用requests访问哔哩哔哩

```python
response=requests.get(url)
data=response.text
print(data)
```

下面还是展示一下怎么在requests中添加headers。在requests中添加headers还是很简单的，只需要添加改一个地方

```python
response=requests.get(url,headers=headers) # 添加headers
data=response.text
print(data)
```
